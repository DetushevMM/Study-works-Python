**–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –ü–∞—Å–∫–∞–ª—è 1 üå∂Ô∏è**

–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –ü–∞—Å–∫–∞–ª—è ‚Äì –±–µ—Å–∫–æ–Ω–µ—á–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤, –∏–º–µ—é—â–∞—è —Ç—Ä–µ—É–≥–æ–ª—å–Ω—É—é —Ñ–æ—Ä–º—É. –í —ç—Ç–æ–º —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–µ –Ω–∞ –≤–µ—Ä—à–∏–Ω–µ –∏ –ø–æ –±–æ–∫–∞–º —Å—Ç–æ—è—Ç –µ–¥–∏–Ω–∏—Ü—ã. –ö–∞–∂–¥–æ–µ —á–∏—Å–ª–æ —Ä–∞–≤–Ω–æ —Å—É–º–º–µ –¥–≤—É—Ö —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω—ã—Ö –Ω–∞–¥ –Ω–∏–º —á–∏—Å–µ–ª.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –†–µ—à–µ–Ω–∏–µ —É–¥–æ–±–Ω–æ –æ—Ñ–æ—Ä–º–∏—Ç—å –≤ –≤–∏–¥–µ —Ñ—É–Ω–∫—Ü–∏–∏ pascal(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–æ–º–µ—Ä —Å—Ç—Ä–æ–∫–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â—É—é —Å—Ç—Ä–æ–∫—É —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è.

```python
def pascal(n):
    lst = [1]  # –Ω–∞—á–∞–ª—å–Ω–∞—è —Å—Ç—Ä–æ–∫–∞
    for i in range(n):
        lst = [0] + lst + [0]  # –¥–æ–±–∞–≤–ª—è–µ–º –Ω—É–ª–∏ –ø–æ –±–æ–∫–∞–º –∫ —Ç–µ–∫—É—â–µ–π —Å—Ç—Ä–æ–∫–µ —Å—Ç—Ä–æ–∫–µ
        s = []  # —Ç—É—Ç –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –Ω–æ–≤–∞—è —Å—Ç—Ä–æ–∫–∞
        for j in range(len(lst) - 1):
            s.append(lst[j] + lst[j + 1])
        lst = s  # –Ω–æ–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –Ω–∞—à–µ–π —Ç–µ–∫—É—â–µ–π —Å—Ç—Ä–æ–∫–æ–π
    return lst


n = int(input())  # 3
print(pascal(n))
```

**–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –ü–∞—Å–∫–∞–ª—è 2**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –ø–µ—Ä–≤—ã–µ n —Å—Ç—Ä–æ–∫ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è.

```python
def pascal(n):
    s = [[1]]  # —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∞—è —Ç–∞–±–ª–∏—Ü–∞
    st = [1]  # –Ω–∞—á–∞–ª—å–Ω–∞—è —Å—Ç—Ä–æ–∫–∞
    for i in range(n - 1):
        st = [0] + st + [0]  # –¥–æ–±–∞–≤–ª—è–µ–º –Ω—É–ª–∏ –ø–æ –±–æ–∫–∞–º –∫ —Ç–µ–∫—É—â–µ–π —Å—Ç—Ä–æ–∫–µ —Å—Ç—Ä–æ–∫–µ
        final_s = []  # —Ç—É—Ç –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –Ω–æ–≤–∞—è —Å—Ç—Ä–æ–∫–∞
        for j in range(len(st) - 1):
            final_s.append(st[j] + st[j + 1])  # –¥–æ–±–∞–≤–ª—è–µ–º –≤ –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É —Å—É–º–º—É —Å–æ—Å–µ–¥–Ω–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Ç–µ–∫—É—â–µ–π —Å—Ç—Ä–æ–∫–∏
        st = final_s  # —Ç–µ–ø–µ—Ä—å –Ω–æ–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –Ω–∞—à–µ–π —Ç–µ–∫—É—â–µ–π —Å—Ç—Ä–æ–∫–æ–π
        s.append(st)  # –¥–æ–±–∞–≤–ª—è–µ–º —Ç–µ–∫—É—â—É—é —Å—Ç—Ä–æ–∫—É –≤ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â—É—é —Ç–∞–±–ª–∏—Ü—É
    return s

n = int(input())  # 4

for k in pascal(n):
    print(*k)  # –≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É –ø–æ —Å—Ç—Ä–æ–∫–∞–º
```

**–£–ø–∞–∫–æ–≤–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ üå∂Ô∏è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è —Å–∏–º–≤–æ–ª—ã. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —É–ø–∞–∫–æ–≤—ã–≤–∞–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –∑–∞–¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ –≤ –ø–æ–¥—Å–ø–∏—Å–∫–∏.

```python
s = input().split()  # a b c d
res = [[s[0]]]
for i in range(1, len(s)):
    if s[i] == s[i - 1]:
        res[-1].append(s[i])
    else:
        res.append([s[i]])
print(res) 
```

**–†–∞–∑–±–∏–µ–Ω–∏–µ –Ω–∞ —á–∞–Ω–∫–∏ üå∂Ô∏è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏: –Ω–∞ –æ–¥–Ω–æ–π ‚Äì —Å–∏–º–≤–æ–ª—ã, –Ω–∞ –¥—Ä—É–≥–æ–π ‚Äì —á–∏—Å–ª–æ n. –ò–∑ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–∏ —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫.

–†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é chunked(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Å–ø–∏—Å–æ–∫ –∏ —á–∏—Å–ª–æ, –∑–∞–¥–∞—é—â–µ–µ —Ä–∞–∑–º–µ—Ä —á–∞–Ω–∫–∞ (–∫—É—Å–∫–∞), –∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –∏–∑ —á–∞–Ω–∫–æ–≤ (–∫—É—Å–∫–æ–≤) —É–∫–∞–∑–∞–Ω–Ω–æ–π –¥–ª–∏–Ω—ã.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ù–µ –∑–∞–±—É–¥—å—Ç–µ –≤—ã–∑–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é chunked(), —á—Ç–æ–±—ã –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç. üòÄ

```python
def chunked(st, n):
    st = st.split()
    a = [[] for _ in range(0, len(st), n)]
    for i in range(len(a)):
        a[i].extend(st[:n])
        st = st[n:]
    return a

string = input()  # a b c d e f
num = int(input())  # 2

print(chunked(string, num))
```

**–ü–æ–¥—Å–ø–∏—Å–∫–∏ —Å–ø–∏—Å–∫–∞ üå∂Ô∏èüå∂Ô∏è**

–ü–æ–¥—Å–ø–∏—Å–æ–∫ ‚Äì —á–∞—Å—Ç—å –¥—Ä—É–≥–æ–≥–æ —Å–ø–∏—Å–∫–∞. –ü–æ–¥—Å–ø–∏—Å–æ–∫ –º–æ–∂–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å –æ–¥–∏–Ω —ç–ª–µ–º–µ–Ω—Ç, –Ω–µ—Å–∫–æ–ª—å–∫–æ –∏–ª–∏ –¥–∞–∂–µ –Ω–∏ –æ–¥–Ω–æ–≥–æ. –ù–∞–ø—Ä–∏–º–µ—Ä, [1], [2], [3] –∏ [4] ‚Äì –ø–æ–¥—Å–ø–∏—Å–∫–∏ —Å–ø–∏—Å–∫–∞ [1, 2, 3, 4]. –°–ø–∏—Å–æ–∫ [2, 3] ‚Äì –ø–æ–¥—Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫–∞ [1, 2, 3, 4], –Ω–æ —Å–ø–∏—Å–æ–∫ [2, 4] –Ω–µ –ø–æ–¥—Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫–∞ [1, 2, 3, 4], —Ç–∞–∫ –∫–∞–∫ —ç–ª–µ–º–µ–Ω—Ç—ã 2 –∏ 4 –≤–æ –≤—Ç–æ—Ä–æ–º —Å–ø–∏—Å–∫–µ –Ω–µ —Å–º–µ–∂–Ω—ã–µ (–æ–Ω–∏ —Ä–∞–∑—Ä—ã–≤–∞—é—Ç—Å—è —ç–ª–µ–º–µ–Ω—Ç–æ–º 3). –ü—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ ‚Äì –ø–æ–¥—Å–ø–∏—Å–æ–∫ –ª—é–±–æ–≥–æ —Å–ø–∏—Å–∫–∞. –°–∞–º —Å–ø–∏—Å–æ–∫ ‚Äì –ø–æ–¥—Å–ø–∏—Å–æ–∫ —Å–∞–º–æ–≥–æ —Å–µ–±—è, —Ç–æ –µ—Å—Ç—å —Å–ø–∏—Å–æ–∫ [1, 2, 3, 4] –ø–æ–¥—Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫–∞ [1, 2, 3, 4].

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è —Å–∏–º–≤–æ–ª—ã. –ò–∑ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç —Å–ø–∏—Å–æ–∫, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –ø–æ–¥—Å–ø–∏—Å–∫–∏ —Å–ø–∏—Å–∫–∞, –≤–∫–ª—é—á–∞—è –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ü–æ—Ä—è–¥–æ–∫ —Å–ø–∏—Å–∫–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –¥–ª–∏–Ω—ã –¥–æ–ª–∂–µ–Ω —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –ø–æ—Ä—è–¥–∫—É –∏—Ö –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –æ—Å–Ω–æ–≤–Ω–æ–π —Å–ø–∏—Å–æ–∫.

```python
n = input().split()  # a b
o = [[]]
k = 1
while k != len(n) + 1:
    for j in range(len(n)):
        if len(n[j:j+k])==k:
            o.append(n[j:j+k])
    k+=1
print (o)
```

**–í—ã–≤–µ—Å—Ç–∏ –º–∞—Ç—Ä–∏—Ü—É 2**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏  m, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ. –î–∞–ª–µ–µ –≤–≤–æ–¥—è—Ç—Å—è —Å–∞–º–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã ‚Äì —Å–ª–æ–≤–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ; –ø–æ–¥—Ä—è–¥ –∏–¥—É—Ç —ç–ª–µ–º–µ–Ω—Ç—ã —Å–Ω–∞—á–∞–ª–∞ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–∏, –∑–∞—Ç–µ–º –≤—Ç–æ—Ä–æ–π, –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã –æ–¥–∏–Ω –∑–∞ –¥—Ä—É–≥–∏–º, –≤—ã–≤–æ–¥–∏—Ç –∏—Ö –≤ –≤–∏–¥–µ –º–∞—Ç—Ä–∏—Ü—ã, –≤—ã–≤–æ–¥–∏—Ç –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É, –∏ —Å–Ω–æ–≤–∞ —Ç—É –∂–µ –º–∞—Ç—Ä–∏—Ü—É, –Ω–æ —É–∂–µ –ø–æ–º–µ–Ω—è–≤ –º–µ—Å—Ç–∞–º–∏ —Å—Ç—Ä–æ–∫–∏ —Å–æ —Å—Ç–æ–ª–±—Ü–∞–º–∏: –ø–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ –≤—ã–≤–æ–¥–∏—Ç—Å—è –∫–∞–∫ –ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü, –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ.

```python
n, m = int(input()), int(input())  # n - —Å—Ç—Ä–æ–∫–∏, m - —Å—Ç–æ–ª–±—Ü—ã
matrix = [[input() for _ in range(m)] for i in range(n)]  # —Å—Ä–∞–∑—É —Å–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É (–ø—Ä–µ–¥—ã–¥—É—â–∞—è –∑–∞–¥–∞—á–∞ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É). ([[–ø—Ä–æ—Ö–æ–¥–∏–º—Å—è input –ø–æ —Å—Ç—Ä–æ–∫–∞–º n] –¥–æ–±–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —á–∞—Å—Ç—è–º, —Å–æ–≥–ª–∞—Å–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç–æ–ª–±—Ü–æ–≤])
 
for i in matrix:  # i - —ç—Ç–æ –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞
    print(*i)  # –≤—ã–≤–æ–¥–∏–º –∫–∞–∂–¥—É—é —Å—Ç—Ä–æ–∫—É –º–∞—Ç—Ä–∏—Ü—ã

print()  # –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø—Ä–æ—Ü–µ—Å—Å –≤—ã–≤–æ–¥–∞ —Å—Ç—Ä–æ–∫ –∏–∑ –º–∞—Ç—Ä–∏—Ü—ã

for j in range(m):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º –º–∞—Ç—Ä–∏—Ü—ã
    for k in range(n):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –ø–æ —Å—Ç—Ä–æ–∫–∞–º n
        print(matrix[k][j], end=' ')  # –≤—ã–≤–æ–¥–∏–º –≤—Å–µ k - —ç—Ç–æ —ç–ª–µ–º–µ–Ω—Ç —Å—Ç—Ä–æ–∫–∏,  j —ç—Ç–æ —ç–ª–µ–º–µ–Ω—Ç —Å—Ç–æ–ª–±—Ü–∞ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª. –°–ø–µ—Ä–≤–∞ –ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü m, –∑–∞—Ç–µ–º –≤—Ç–æ—Ä–æ–π —Å—Ç–æ–ª–±–µ—Ü
    print()  # –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø—Ä–æ—Ü–µ—Å—Å –≤—ã–≤–æ–¥–∞ —Å—Ç—Ä–æ–∫ –∏–∑ –º–∞—Ç—Ä–∏—Ü—ã
```

**–°–ª–µ–¥ –º–∞—Ç—Ä–∏—Ü—ã ‚ÜòÔ∏è**

–°–ª–µ–¥–æ–º –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—É–º–º–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç —Å–ª–µ–¥ –∑–∞–¥–∞–Ω–Ω–æ–π –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã.

```python
cnt_sum = 0  # –∫–∞–∫ –±—ã —Å—á–µ—Ç—á–∏–∫, –∫–æ—Ç–æ—Ä—ã–π —Å—É–º–º–∏—Ä—É–µ—Ç —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫–∏
for i in range(int(input())):
    cnt_sum += int(input().split()[i])  # —É –Ω–∞—Å i —ç—Ç–æ —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫–∏ (–Ω–∞—á–∏–Ω–∞—è —Å 1 –∏ –¥–æ n), –∞ —Å—Ç—Ä–æ–∫ –∏  —Å—Ç–æ–ª–±—Ü–æ–≤ —É –Ω–∞—Å –æ—Ç 1 –¥–æ n = int(input()). –ü–æ–ª—É—á–∞–µ—Ç—Å—è –º—ã –≤—ã–≤–æ–¥–∏–º —Å–ø–µ—Ä–≤–∞ 1 —ç–ª–µ–º–µ–Ω—Ç —Å—Ç—Ä–æ–∫–∏, –ø–æ—Ç–æ–º 2 –∏ —Ç–¥. –ò –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç i –º—ã —Å—É–º–º–∏—Ä—É–µ–º —Å–æ —Å–ª–µ–¥—É—é—â–∏–º i, –ø–µ—Ä–µ–≤–µ–¥—è –≤ int –∏ —Ä–∞–∑–¥–µ–ª–∏–≤ —Å—Ç—Ä–æ–∫—É –Ω–∞ —ç–ª–µ–º–µ–Ω—Ç—ã —Å –ø–æ–º–æ—â—å—é split()
print(cnt_sum)
```

**–ë–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ, –±–æ–ª—å—à–∏—Ö —Å—Ä–µ–¥–Ω–µ–≥–æ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏.

```python
n = int(input())  
cnt = 0  # –¥–µ–ª–∞–µ–º —Å—á–µ—Ç—á–∏–∫
ls = [input().split() for i in range(n)]  # —Å–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –∏ —Å—Ä–∞–∑—É –µ–≥–æ –∑–∞–≥—Ä—É–∂–∞–µ–º –≤ –º–∞—Ç—Ä–∏—Ü—É –∏ —Ä–∞–∑–±–∏–≤–∞–µ–º –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º
for j in ls:  #  –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –∫–∞–∂–¥—É—é —Å—Ç—Ä–æ–∫—É –≤ —Å–ø–∏—Å–∫–µ
    sred = (sum(list(map(int, j)))) / len(j)  # –≤—ã—á–∏—Å–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ = (—Å—É–º–º–∞(—Å–ø–∏—Å–æ–∫(–æ—Ç–æ–±—Ä–∞–∑–∏—Ç—å(—á–∏—Å–ª–æ —ç–ª–µ–º–µ–Ω—Ç–∞)))) / –¥–µ–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤. —Å—É–º–º–∏—Ä—É–µ–º(list(map(int - —Å –ø–æ–º–æ—â—å—é map –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º list –≤ int —É–∫–∞–∑–∞–Ω–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç j)))
    for k in j:  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ –∫–∞–∂–¥–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å—Ç—Ä–æ–∫–∏
        if int(k) > sred:  # –≤—ã—è—Å–Ω—è–µ–º, –±–æ–ª—å—à–µ –ª–∏ —ç–ª–µ–º–µ–Ω—Ç —á–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã —ç–ª–µ–º–µ–Ω—Ç–æ–≤
            cnt += 1  # –µ—Å–ª–∏ –¥–∞, —Ç–æ +1 –∫ —Å—á–µ—Ç—á–∏–∫—É
    print(cnt)  # —Å—Ä–∞–∑—É –ø–µ—á–∞—Ç–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏
    cnt = 0  # –æ–±–Ω—É–ª—è–µ–º —Å—á–µ—Ç—á–∏–∫, —á—Ç–æ–± –ø–æ—Å—á–∏—Ç–∞—Ç—å –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É –∏ —Ç.–¥. 
```

**–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –≤ –æ–±–ª–∞—Å—Ç–∏ 1**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ –∑–∞—à—Ç—Ä–∏—Ö–æ–≤–∞–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã (—Ç—Ä–µ—Ü–≥–æ–ª—å–Ω–∏–∫)

```python
n = int(input())
lst = []  # –ø—É—Å—Ç–æ–π –ª–∏—Å—Ç –¥–ª—è –∏—Ç–æ–≥–∞
mtx = [[int(i) for i in input().split()] for _ in range(n)]  # [[–ø–µ—Ä–µ–≤–æ–¥–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —á–∏—Å–ª–∞ –≤ –≤ –≤–æ–¥–∏–º–æ–π —Å—Ç—Ä–æ–∫–µ, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω–æ–π —Å–ø–ª–∏—Ç–æ–º] –ø–µ—Ä–µ—Ö–æ–¥–∏–º –Ω–∞ —Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É]
for i in range(n):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å—Ç—Ä–æ–∫–∞–º–∏
    for j in range(n): # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º
        if i >= j:  # –ø–æ–ª—É—á–∞–µ—Ç—Å—è —Ç–∞–∫: 1 —á–∏—Å–ª–æ —Å—Ç–æ–ª–±—Ü–∞ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç—Å—è —Å–æ 2 —á–∏—Å–ª–æ–º —Å—Ç—Ä–æ–∫–∏, 2 —Å —Ç—Ä–µ—Ç—å–∏–º, —Ç—Ä–µ—Ç—å–µ —Å –Ω–∏—á–µ–º –∏ —Ç–¥. –ü–æ–ª—É—á–∞–µ—Ç—Å—è –¥–∏–∞–≥–æ–Ω–∞–ª—å. –í—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ —Å—Ç–æ–ª–±—Ü–∞ ... —Ö—Ä–µ–Ω –∑–Ω–∞–µ—Ç...
            lst.append(mtx[i][j])
print(max(lst))
```

**–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –≤ –æ–±–ª–∞—Å—Ç–∏ 2 üå∂Ô∏è**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ –∑–∞—à—Ç—Ä–∏—Ö–æ–≤–∞–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã (><)/ –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –≠–ª–µ–º–µ–Ω—Ç—ã –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π —Ç–∞–∫–∂–µ —É—á–∏—Ç—ã–≤–∞—é—Ç—Å—è.

```python
n = int(input())
m_el = []  # —Å–æ–∑–¥–∞–µ–º –∏—Ç–æ–≥–æ–≤—ã–π —Å–ø–∏—Å–æ–∫, –∫—É–¥–∞ –∑–∞—Å—É–Ω–µ–º –≤—Å–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏
for i in range(n):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å—Ç—Ä–æ–∫–∞–º –≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –Ω —Ä–∞–∑ (–∏ - —ç—Ç–æ 1 —Å—Ç—Ä–æ–∫–∞)
    mtx = [int(i) for i in input().split()]  # –ø–µ—Ä–µ–≤–æ–¥–∏–º –∫–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª —Å—Ç—Ä–æ–∫–∏ (–≤–≤–æ–¥–∏—Ç —á–µ—Ä–µ–∑ –∏–Ω–ø—É—Ç) –≤ —á–∏—Å–ª–∞ –∏ —Å–æ–∑–¥–∞–µ–º —á–∏—Å–ª–µ–Ω–Ω—É—é (–∏–Ω—Ç) —Å—Ç—Ä–æ–∫—É, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—É—é —Å–ø–ª–∏—Ç–æ–º
    for j in range(len(mtx)):  # —ç—Ç–æ —Ü–∏–∫–ª,–æ–Ω —Ä–∞–±–æ—Ç–∞–µ—Ç —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ —Å–ø–∏–∫–∞ —Å —á–∏—Å–ª–∞–º–∏. –ë–µ—Ä–µ–º 1 —á–∏—Å–ª–æ –∏–∑ —Å–ø–∏—Å–∫–∞(–¥–ª—è —ç—Ç–æ–≥–æ –¥–µ–ª–∞–µ–º –ª–µ–Ω)
        if (i >= j and i <= n - 1 -j) or (i <= j and i >= n - 1 -j) or (i == j) or (i + j + 1 == n):  # —ç—Ç–æ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã –≤ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö —Å–ø–µ—Ä–≤–∞ –ª–µ–≤—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫, –ø–æ—Ç–æ–º –∏–ª–∏ –ø—Ä–∞–≤—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫, –ø–æ—Ç–æ–º –∏–ª–∏ –æ–Ω–∏ —Ä–∞–≤–Ω—ã
            m_el.append(mtx[j])  # –¥–æ–±–∞–≤–ª—è–µ–º —Ç–∞–∫–æ–µ –¥–∂–∏ –∏–∑ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –∏ –∫ –∏—Ç–æ–≥–æ–≤–æ–º—É —Å–ø–∏—Å–∫—É –∏ –ø–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–π —Å—Ç—Ä–æ–∫–µ
print(max(m_el))  # –ø–µ—á–∞—Ç–∞–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —Ü–∏—Ñ—Ä—É –∏–∑ –º–∞—Ç–∏—Ä–∏—Ü—ã
```

**–°—É–º–º—ã —á–µ—Ç–≤–µ—Ä—Ç–µ–π**

–ö–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Ä–∞–∑–±–∏–≤–∞–µ—Ç—Å—è –Ω–∞ —á–µ—Ç—ã—Ä–µ —á–µ—Ç–≤–µ—Ä—Ç–∏, –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω—ã–µ –≥–ª–∞–≤–Ω–æ–π –∏ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª—è–º–∏: –≤–µ—Ä—Ö–Ω—é—é, –Ω–∏–∂–Ω—é—é, –ª–µ–≤—É—é –∏ –ø—Ä–∞–≤—É—é. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç —Å—É–º–º—É —ç–ª–µ–º–µ–Ω—Ç–æ–≤: –≤–µ—Ä—Ö–Ω–µ–π —á–µ—Ç–≤–µ—Ä—Ç–∏; –ø—Ä–∞–≤–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏; –Ω–∏–∂–Ω–µ–π —á–µ—Ç–≤–µ—Ä—Ç–∏; –ª–µ–≤–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –≠–ª–µ–º–µ–Ω—Ç—ã –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π –Ω–µ —É—á–∏—Ç—ã–≤–∞—é—Ç—Å—è.

```python
n = int(input())  # —Å—Ç–æ—Ä–æ–Ω—ã –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã
mtx = [[int(i) for i in input().split()] for _ in range(n)]  # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Ç—Ä–∏—Ü—É
left = 0  # —Å–æ–∑–¥–∞–µ–º —á–∏—Å–ª–µ–Ω–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –Ω–∞ –∫–∞–∂–¥—É—é —á–µ—Ç–≤–µ—Ä—Ç—å
right = 0  
up = 0  
down = 0  
for i in range(n):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å—Ç—Ä–æ–∫–∞–º
    for j in range(n):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å—Ç—Ä–æ–∫ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ
        if i < j and i < n - j - 1:  # –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –≤–µ—Ä—Ö–Ω–µ–π —á–µ—Ç–≤–µ—Ä—Ç–∏ (–±–µ–∑ —É—á–µ—Ç–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π (—É–±—Ä–∞–ª–∏ = –∏–∑ —É—Å–ª–æ–≤–∏—è))
            up += mtx[i][j]  # —Å—É–º–º–∏—Ä—É–µ–º –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å–æ–≥–ª–∞—Å–Ω–æ —É—Å–ª–æ–≤–∏—é
        if i < j and i > n - j - 1:  # –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ø—Ä–∞–≤–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏
            right += mtx[i][j]  # —Å—É–º–º–∏—Ä—É–µ–º –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã
        if i > j and i > n - j - 1:  # –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –Ω–∏–∂–Ω–µ–π —á–µ—Ç–≤–µ—Ä—Ç–∏
            down += mtx[i][j]   # —Å—É–º–º–∏—Ä—É–µ–º –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã
        if i > j and i < n - j - 1:  # –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ª–µ–≤–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏
            left += mtx[i][j]  # —Å—É–º–º–∏—Ä—É–µ–º –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã
print('–í–µ—Ä—Ö–Ω—è—è —á–µ—Ç–≤–µ—Ä—Ç—å:', up)
print('–ü—Ä–∞–≤–∞—è —á–µ—Ç–≤–µ—Ä—Ç—å:', right)
print('–ù–∏–∂–Ω—è—è —á–µ—Ç–≤–µ—Ä—Ç—å:', down)
print('–õ–µ–≤–∞—è —á–µ—Ç–≤–µ—Ä—Ç—å:',left)  
```

**–¢–∞–±–ª–∏—Ü–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ. –°–æ–∑–¥–∞–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—É mult —Ä–∞–∑–º–µ—Ä–æ–º n√óm –∏ –∑–∞–ø–æ–ª–Ω–∏—Ç–µ –µ–µ —Ç–∞–±–ª–∏—Ü–µ–π —É–º–Ω–æ–∂–µ–Ω–∏—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ mult[i][j] = i * j.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–∞–±–ª–∏—Ü—É —É–º–Ω–æ–∂–µ–Ω–∏—è –æ—Ç–≤–æ–¥—è –Ω–∞ –≤—ã–≤–æ–¥ –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ (–¥–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust()).

```python
n, m = int(input()), int(input())  # –¥–ª–∏–Ω–Ω–∞ –∏ —à–∏—Ä–∏–Ω–∞  4 –∏ 6
mult = [[str((i * j)).ljust(3) for i in range(m)] for j in range(n)]  # —Å–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É –ø–æ —Å—Ç—Ä–æ–∫–∞–º –≤–º–µ—Å—Ç–æ —Ü–∏—Ñ—Ä –ø–æ —Ñ–æ—Ä–º—É–ª–µ i * j, –æ—Ç–≤–æ–¥—è –Ω–∞ –≤—ã–≤–æ–¥ –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ (ljust())
for k in mult:  # –∏–∑–≤–ª–µ–∫–∞–µ–º —Å–ø–∏—Å–∫–∏ –∏–∑ —Å–ø–∏–∫–∞ –º—É–ª—å—Ç –ø–æ –æ–¥–Ω–æ–º—É
    print(*k)  # –≤—ã–≤–æ–¥–∏–º —ç–ª–µ–º–µ–Ω—Ç—ã –∫–∞–∂–¥–æ–≥–æ —Å–ø–∏—Å–∫–∞ –ø–æ —Å—Ç—Ä–æ–∫–∞–º, –æ–¥–Ω–∞ –∑–∞ –¥—Ä—É–≥–æ–π, –≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ m —Å—Ç–æ–±—Ü–æ–≤
```

**–ú–∞–∫—Å–∏–º—É–º –≤ —Ç–∞–±–ª–∏—Ü–µ üîù**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ, –∑–∞—Ç–µ–º n —Å—Ç—Ä–æ–∫ –ø–æ m —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –≤ –∫–∞–∂–¥–æ–π, –æ—Ç–¥–µ–ª–µ–Ω–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –∏–Ω–¥–µ–∫—Å—ã (—Å—Ç—Ä–æ–∫—É –∏ —Å—Ç–æ–ª–±–µ—Ü) –ø–µ—Ä–≤–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –°—á–∏—Ç–∞–π—Ç–µ, —á—Ç–æ –Ω—É–º–µ—Ä–∞—Ü–∏—è —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –Ω—É–ª—è.

```python
n, m = int(input()), int(input())
matrix = [[int(i) for i in input().split()] for j in range(n)]
row, col = 0, 0

for i in range(n):
    for j in range(m):
        if matrix[i][j] > matrix[row][col]:
            row,col = i, j
            
print(row, col) 
```

**–û–±–º–µ–Ω —Å—Ç–æ–ª–±—Ü–æ–≤ ‚è∏Ô∏è**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –º–µ–Ω—è–µ—Ç –º–µ—Å—Ç–∞–º–∏ —Å—Ç–æ–ª–±—Ü—ã –≤ –º–∞—Ç—Ä–∏—Ü–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Å—Ç—Ä–æ–∫–∞—Ö –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ, –∑–∞—Ç–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã –ø–æ—Å—Ç—Ä–æ—á–Ω–æ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª, –∑–∞—Ç–µ–º –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ i –∏ j ‚Äì –Ω–æ–º–µ—Ä–∞ —Å—Ç–æ–ª–±—Ü–æ–≤, –ø–æ–¥–ª–µ–∂–∞—â–∏—Ö –æ–±–º–µ–Ω—É.

```python
n, m = int(input()), int(input())  # –¥–ª–∏–Ω–∞ –∏ —à–∏—Ä–∏–Ω–∞ –º–∞—Ç—Ä–∏—Ü—ã
mtx = [[int(i) for i in input().split()] for i in range(n)]  # —Å–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É —Å —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ - —á–∏—Å–ª–∞–º–∏ –ø–æ—Å—Ç—Ä–æ—á–Ω–æ
change = [int(j) for j in input().split()]  # –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ–¥–∞—é—Ç—Å—è –Ω–æ–º–µ—Ä–∞ —Å—Ç–æ–ª–±—Ü–æ–≤, —Å–æ–∑–¥–∞–µ–º —á–∏—Å–ª–æ–≤—É—é –º–∞—Ç—Ä–∏—Ü—É –∏–∑ –Ω–∏—Ö [i, j]
i = change[0]  #  i –∏ j ‚Äî –Ω–æ–º–µ—Ä–∞ —Å—Ç–æ–ª–±—Ü–æ–≤, –ø–æ–¥–ª–µ–∂–∞—â–∏—Ö –æ–±–º–µ–Ω—É.
j = change[1]  
for k in range(n):  # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è —Å—Ç—Ä–æ–∫–∞–º–∏ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º
    mtx[k][i], mtx[k][j] = mtx[k][j], mtx[k][i]  # –º–µ–Ω—è–µ–º –º–µ—Å—Ç–∞–º–∏ —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫ –ø–æ –Ω–æ–º–µ—Ä—É i –∏ –ø–æ –Ω–æ–º–µ—Ä—É j
for g in mtx:  # –≤—ã–≤–æ–¥–∏–º –∫–∞–∂–¥—É—é —Å—Ç—Ä–æ–∫—É –∏–∑ —Å–ø–∏—Å–∫–∞ –ø–æ–æ—á–µ—Ä–µ–¥–Ω–æ
    print(*g)    
```

**–°–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ üí†**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ—Å—Ç—å –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ, –∑–∞—Ç–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã –ø–æ—Å—Ç—Ä–æ—á–Ω–æ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ YES, –µ—Å–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏, –∏–ª–∏ NO –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)] 
flag = 'YES'
for i in range(n):
    for j in range(n):
        if i != j:
            if mtx[i][j] != mtx[j][i]:
                flag = 'NO'
                break
print(flag)
```

**–û–±–º–µ–Ω –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π üîÉ**

–î–∞–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —á–∏—Å–µ–ª. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –º–µ–Ω—è–µ—Ç –º–µ—Å—Ç–∞–º–∏ —ç–ª–µ–º–µ–Ω—Ç—ã, —Å—Ç–æ—è—â–∏–µ –Ω–∞ –≥–ª–∞–≤–Ω–æ–π –∏ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏, –ø—Ä–∏ —ç—Ç–æ–º –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç –¥–æ–ª–∂–µ–Ω –æ—Å—Ç–∞—Ç—å—Å—è –≤ —Ç–æ–º –∂–µ —Å—Ç–æ–ª–±—Ü–µ (—Ç–æ –µ—Å—Ç—å –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ –Ω—É–∂–Ω–æ –ø–æ–º–µ–Ω—è—Ç—å –º–µ—Å—Ç–∞–º–∏ —ç–ª–µ–º–µ–Ω—Ç –Ω–∞ –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –∏ –Ω–∞ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏).

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)]  #
for i in range(n):  # 
    for j in range(n):  # 
        if (i == j) or (i == n - 1 - j):  # –µ—Å–ª–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –Ω–∞—Ö–æ–¥—è—Ç—Å—è –Ω–∞ –≥–ª–∞–≤–Ω–æ–π –∏–ª–∏ –æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏
            print(mtx[n - 1 - i][j], end=' ')  # –º–µ–Ω—è–µ–º –º–µ—Å—Ç–∞–º–∏ –∏ –≤—ã–≤–æ–¥–∏–º
        else:
            print(mtx[i][j], end=' ')  # –æ—Å—Ç–∞–≤–ª—è–µ–º –∫–∞–∫ –µ—Å—Ç—å
    print()
```

**–ó–µ—Ä–∫–∞–ª—å–Ω–æ–µ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ ü¶ã**

–î–∞–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —á–∏—Å–µ–ª. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –∑–µ—Ä–∫–∞–ª—å–Ω–æ –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –µ—ë —ç–ª–µ–º–µ–Ω—Ç—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ —Å–∏–º–º–µ—Ç—Ä–∏–∏.

```python
n = int(input())
mtx = [input().split() for _ in range(n)]  # —Å–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É —Ü–µ–ª—ã–º–∏ —Å—Ç—Ä–æ–∫–∞–º–∏
for i in range(n // 2):  # –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å—Ç—Ä–æ–∫–∞–º –¥–æ —Å–µ—Ä–µ–¥–∏–Ω—ã –º–∞—Ç—Ä–∏—Ü—ã
    mtx[i], mtx[n - 1 - i] = mtx[n - 1 - i], mtx[i]  # –º–µ–Ω—è–µ–º –º–µ—Å—Ç–∞–º–∏ —Å—Ç—Ä–æ–∫–∏
for g in mtx:
    print(*g)
```

**–ü–æ–≤–æ—Ä–æ—Ç –º–∞—Ç—Ä–∏—Ü—ã üîÅ**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É —á–∏—Å–µ–ª –Ω–∞ 90‚àò –ø–æ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–µ.

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)]
for i in range(n):
    for j in range(n - 1, -1, -1):
        print(mtx[j][i], end=' ')
    print()  
```

**–•–æ–¥—ã –∫–æ–Ω—è üêé**
–ù–∞ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–µ 8√ó8 —Å—Ç–æ–∏—Ç –∫–æ–Ω—å. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ—Ç–º–µ—á–∞–µ—Ç –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∫–æ–Ω—è –Ω–∞ –¥–æ—Å–∫–µ –∏ –≤—Å–µ –∫–ª–µ—Ç–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—å–µ—Ç –∫–æ–Ω—å. –ö–ª–µ—Ç–∫—É, –≥–¥–µ —Å—Ç–æ–∏—Ç –∫–æ–Ω—å, –æ—Ç–º–µ—Ç—å—Ç–µ –∞–Ω–≥–ª–∏–π—Å–∫–æ–π –±—É–∫–≤–æ–π N, –∞ –∫–ª–µ—Ç–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—å–µ—Ç –∫–æ–Ω—å, –æ—Ç–º–µ—Ç—å—Ç–µ —Å–∏–º–≤–æ–ª–∞–º–∏ *, –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ –∑–∞–ø–æ–ª–Ω–∏—Ç–µ —Ç–æ—á–∫–∞–º–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫–æ–Ω—è –Ω–∞ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–µ –≤ —à–∞—Ö–º–∞—Ç–Ω–æ–π –Ω–æ—Ç–∞—Ü–∏–∏ (—Ç–æ –µ—Å—Ç—å –≤ –≤–∏–¥–µ e4, –≥–¥–µ —Å–Ω–∞—á–∞–ª–∞ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –Ω–æ–º–µ—Ä —Å—Ç–æ–ª–±—Ü–∞ (–±—É–∫–≤–∞ –æ—Ç a –¥–æ h, —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ), –∑–∞—Ç–µ–º –Ω–æ–º–µ—Ä—É —Å—Ç—Ä–æ–∫–∏ (—Ü–∏—Ñ—Ä–∞ –æ—Ç 1 –¥–æ 8, —Å–Ω–∏–∑—É –≤–≤–µ—Ä—Ö)).

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –¥–æ—Å–∫–∏, —Ä–∞–∑–¥–µ–ª—è—è —ç–ª–µ–º–µ–Ω—Ç—ã –ø—Ä–æ–±–µ–ª–∞–º–∏.

```python
a, b = input()  # b6
n = 8
board = [['.'] * n for _ in range(n)]
a = ord(a) - 97  # Ord() –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–º–µ—Ä –∫–æ–¥–æ–≤–æ–π —Ç–æ—á–∫–∏ Unicode –ø–µ—Ä–≤–æ–≥–æ —Å–∏–º–≤–æ–ª–∞ –≤—Ö–æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏
b = n - int(b)
board[b][a] = 'N'
for i in range(n):
    for j in range(n):
        if abs(b - i) * abs(a - j) == 2:
            board[i][j] = '*'
for g in board:
    print(*g)
```

**–ú–∞–≥–∏—á–µ—Å–∫–∏–π –∫–≤–∞–¥—Ä–∞—Ç ‚ú®üå∂Ô∏è**

–ú–∞–≥–∏—á–µ—Å–∫–∏–º –∫–≤–∞–¥—Ä–∞—Ç–æ–º –ø–æ—Ä—è–¥–∫–∞ n –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Ä–∞–∑–º–µ—Ä–∞ n√ón, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–∞—è –∏–∑ –≤—Å–µ—Ö —á–∏—Å–µ–ª 1,2,3, ‚Ä¶, n^2 —Ç–∞–∫, —á—Ç–æ —Å—É–º–º—ã –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É, –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –∏ –∫–∞–∂–¥–æ–π –∏–∑ –¥–≤—É—Ö –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π —Ä–∞–≤–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –º–∞–≥–∏—á–µ—Å–∫–∏–º –∫–≤–∞–¥—Ä–∞—Ç–æ–º.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ, –∑–∞—Ç–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã: n —Å—Ç—Ä–æ–∫, –ø–æ n —á–∏—Å–µ–ª –≤ –∫–∞–∂–¥–æ–π, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ –ø—Ä–æ–±–µ–ª–∞–º–∏.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ YES, –µ—Å–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞ —è–≤–ª—è–µ—Ç—Å—è –º–∞–≥–∏—á–µ—Å–∫–∏–º –∫–≤–∞–¥—Ä–∞—Ç–æ–º, –∏–ª–∏ NO –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)] 
turn_90 = [[[] for i in range(n)] for i in range(n)]  # –ø–æ–≤–µ—Ä–Ω—É—Ç–∞—è –Ω–∞ 90¬∞ –º–∞—Ç—Ä–∏—Ü–∞
cnt = 0  #  —Å—á–µ—Ç—á–∏–∫ –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ —Å—Ç—Ä–æ–∫ –Ω–∞—á–∞–ª—å–Ω–æ–π –∏ –ø–æ–≤–µ—Ä–Ω—É—Ç–æ–π –º–∞—Ç—Ä–∏—Ü—ã
main_diag = []  # –¥–ª—è –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏
side_diag = []  # –¥–ª—è –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏
lst = []  # —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –¥–ª—è –∏—Å–∫–ª—é—á–µ–Ω–∏—è –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π –∏ –ø—Ä–æ–≤–µ—Ä–∫–∏ –æ—Ç 1 –¥–æ 1 ** n
for i in range(n):
    main_diag.append(mtx[i][i])  #  –∑–∞–ø–æ–ª–Ω—è–µ–º —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º-–º–∏ –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏
    side_diag.append(mtx[i][n - i - 1])  #  –∑–∞–ø–æ–ª–Ω—è–µ–º —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏
    for j in range(n):
        turn_90[i][j] = mtx[n - j - 1][i]  # –∑–∞–ø–æ–ª–Ω—è–µ–º –ø–æ–≤–µ—Ä–Ω—É—Ç—É—é –Ω–∞ 90 –º–∞—Ç—Ä–∏—Ü—É
        if mtx[i][j] not in lst:
            lst.append(mtx[i][j])  # –∑–∞–ø–æ–ª–Ω—è–µ–º —Å–ø–∏—Å–æ–∫ —É—á–µ—Ç–∞ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
for i in range(n):
    if sum(mtx[i]) == sum(turn_90[i]) == sum(main_diag) == sum(side_diag):
        cnt += 1
print('YES' if cnt == n and len(lst) == n ** 2 and 0 not in lst else 'NO')
```

**–®–∞—Ö–º–∞—Ç–Ω–∞—è –¥–æ—Å–∫–∞**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã —Ä–∞–∑–º–µ—Ä–æ–º n√óm, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ —Å–∏–º–≤–æ–ª–∞–º–∏ . –∏ * –≤ —à–∞—Ö–º–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ. –í –ª–µ–≤–æ–º –≤–µ—Ä—Ö–Ω–µ–º —É–≥–ª—É –¥–æ–ª–∂–Ω–∞ —Å—Ç–æ—è—Ç—å —Ç–æ—á–∫–∞. –í—ã–≤–µ–¥–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –Ω–∞ —ç–∫—Ä–∞–Ω, —Ä–∞–∑–¥–µ–ª—è—è —ç–ª–µ–º–µ–Ω—Ç—ã –ø—Ä–æ–±–µ–ª–∞–º–∏.

```python
n, m = [int(i) for i in input().split()]  # 3 4
mtx = [['.'] * m for _ in range(n)] 
for i in range(n):
    if i == 0 or i % 2 == 0:
        for j in range(1, m, 2):
            mtx[i][j] = '*'
    else:
        for j in range(0, m, 2):
            mtx[i][j] = '*' 
for g in mtx:
    print(*g) 
```

**–ü–æ–±–æ—á–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ón –∏ –∑–∞–ø–æ–ª–Ω—è–µ—Ç –µ–µ –ø–æ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—Ä–∞–≤–∏–ª—É:
- —á–∏—Å–ª–∞ –Ω–∞ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏ —Ä–∞–≤–Ω—ã 1;
- —á–∏—Å–ª–∞, —Å—Ç–æ—è—â–∏–µ –≤—ã—à–µ —ç—Ç–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏, —Ä–∞–≤–Ω—ã 0;
- —á–∏—Å–ª–∞, —Å—Ç–æ—è—â–∏–µ –Ω–∏–∂–µ —ç—Ç–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏, —Ä–∞–≤–Ω—ã 2.

–ü–æ–ª—É—á–µ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω. –ß–∏—Å–ª–∞ –≤ —Å—Ç—Ä–æ–∫–µ —Ä–∞–∑–¥–µ–ª—è–π—Ç–µ –æ–¥–Ω–∏–º –ø—Ä–æ–±–µ–ª–æ–º.

```python
n = int(input())
matrix = [[0] * n for _ in range(n)]

for i in range(n):
    matrix[i][n - 1 - i] = 1
    for j in range(n):
        if i > n - 1 - j:
            matrix[i][j] = 2
        print(matrix[i][j], end = " ")
    print()
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ 1**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√óm –∏ –∑–∞–ø–æ–ª–Ω—è–µ—Ç –µ–µ —á–∏—Å–ª–∞–º–∏ –æ—Ç 1 –¥–æ n‚ãÖm –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –Ω–∞ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –º–∞—Ç—Ä–∏—Ü—É –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è –≤—ã–≤–æ–¥–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Ç–≤–æ–¥–∏—Ç–µ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ –Ω–∞ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–æ –æ–±–æ–π—Ç–∏—Å—å –∏ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–º–µ—Ç –∏ —Ç–∞–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ. üòá

```python
n, m = map(int, input().split())  # 3 4
mtx = []
counter = 1
for i in range(n):
  mtx.append([])
  for j in range(m):
    mtx[i].append(counter)
    counter += 1
    print(f'{mtx[i][j]}'.ljust(3), end=' ')
  print()
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ 2**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√óm, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è –≤—ã–≤–æ–¥–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Ç–≤–æ–¥–∏—Ç–µ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ –Ω–∞ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–æ –æ–±–æ–π—Ç–∏—Å—å –∏ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–º–µ—Ç –∏ —Ç–∞–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ. üòá

```python
s = input().split()  # 3 7
n = int(s[0])
m = int(s[1])
matrix = [[0 for i in range(m)] for _ in range(n)]
count = 1
for i in range(m):
  for j in range(n):
    matrix[j][i] += count
    count += 1
for g in range(n):
  print(*matrix[g])
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ 3**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ón, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è –≤—ã–≤–æ–¥–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Ç–≤–æ–¥–∏—Ç–µ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ –Ω–∞ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–æ –æ–±–æ–π—Ç–∏—Å—å –∏ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–º–µ—Ç –∏ —Ç–∞–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ. üòá

```python
n = int(input())  # 5

for i in range(n):
  mtx = []
  g = n - 1 - i
  for j in range(n):
    if i == j or g == j:
       mtx.append(1)
    else:
       mtx.append(0)
  print(*mtx)   
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ 4**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ón, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è –≤—ã–≤–æ–¥–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Ç–≤–æ–¥–∏—Ç–µ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ –Ω–∞ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–æ –æ–±–æ–π—Ç–∏—Å—å –∏ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–º–µ—Ç –∏ —Ç–∞–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ. üòá

```python
n = int(input())  # 5
for i in range(n):
  mtx = []
  for j in range(n):
    if (i <= j and i <= n - 1 - j) or (i >= j and i >= n - 1 - j):
      mtx.append(1)
    else:
      mtx.append(0)
  print(*mtx)
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ 5 üå∂Ô∏è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√óm, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è –≤—ã–≤–æ–¥–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Ç–≤–æ–¥–∏—Ç–µ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ –Ω–∞ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–æ –æ–±–æ–π—Ç–∏—Å—å –∏ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–º–µ—Ç –∏ —Ç–∞–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ. üòá

```python
n, m = [int(i) for i in input().split()]  # 5 5
mtx = [i for i in range(1, m + 1)]
for i in range(n):
  for j in range(m):
    print(mtx[j], end=' ')
  pop = mtx.pop(0)
  mtx.append(pop)
  print()
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –∑–º–µ–π–∫–æ–π üêç**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√óm, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ "–∑–º–µ–π–∫–æ–π" –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è –≤—ã–≤–æ–¥–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Ç–≤–æ–¥–∏—Ç–µ —Ä–æ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∞ –Ω–∞ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–æ –æ–±–æ–π—Ç–∏—Å—å –∏ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–º–µ—Ç –∏ —Ç–∞–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ. üòá

```python
s = input().split()  # 5 5
n = int(s[0])
m = int(s[1])
mtx = [[0 for i in range(m)] for _ in range(n)]
cnt = 1
for i in range(n):
  for j in range(m):
    mtx[i][j] += cnt
    cnt += 1
for g in range(n):
  if g % 2 == 0:
    print(*mtx[g])
  else:
    mtx[g].reverse()
    print(*mtx[g])
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –¥–∏–∞–≥–æ–Ω–∞–ª—è–º–∏ ‚ÜôÔ∏èüå∂Ô∏è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√óm, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ "–¥–∏–∞–≥–æ–Ω–∞–ª—è–º–∏" –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

```python
s = input().split()  # 5 5
n = int(s[0])
m = int(s[1])
mtx = [[0] * m for _ in range(n)]
test = 1
koef = 0
while test != n * m + 1:
  for i in range(n):
    for j in range(m):
      if i + j == koef:
        mtx[i][j] = test
        test += 1
  koef += 1
for i in range(n):
  for j in range(m):
    print(str(mtx[i][j]).ljust(3), end=' ')
  print()
```

**–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ —Å–ø–∏—Ä–∞–ª—å—é üåÄüå∂Ô∏èüå∂Ô∏è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√óm, –∑–∞–ø–æ–ª–Ω–∏–≤ –µ–µ "—Å–ø–∏—Ä–∞–ª—å—é" –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º.

```python
n, m = [int(i) for i in input().split()]  # 4 5
mtx = [[0] * int(m) for i in range(n)]
rows = n - 1
pillars = m - 1
a, b, c = 0, 0, 1

while a <= pillars and b <= rows:
  for i in range(a, pillars + 1):
    mtx[a][i] = c
    c += 1
  b += 1
  for j in range(b, rows + 1):
    mtx[j][pillars] = c
    c += 1
  pillars -= 1
  if b <= rows:
    for i in range(pillars, a - 1, -1):
      mtx[rows][i] = c
      c += 1
    rows -= 1
  if a <= pillars:
    for i in range(rows, b - 1, -1):
      mtx[i][a] = c
      c += 1
    a += 1
   
for s in mtx:
  print(*s)
```

**–°–ª–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Å—É–º–º—ã –¥–≤—É—Ö –º–∞—Ç—Ä–∏—Ü.

```python
n, m = [int(i) for i in input().split()]  # n, m - —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–±—Ü—ã

A = [[int(i) for i in input().split()] for _ in range(n)]  # –º–∞—Ç—Ä–∏—Ü–∞ 1
a = input()  # –ü–†–û–ë–ï–õ –ú–ï–ñ–î–£ –ú–ê–¢–†–ò–¶–ê–ú–ò –ø—Ä–∏ –≤–≤–æ–¥–µ –¥–∞–Ω–Ω—ã—Ö
B = [[int(i) for i in input().split()] for _ in range(n)]  # –º–∞—Ç—Ä–∏—Ü–∞ 2
res_mtx = [[0] * m for i in range(n)]  # –º–∞—Ç—Ä–∏—Ü–∞, –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–∞—è –Ω—É–ª—è–º–∏

for i in range(n):
    for j in range(m):
        res_mtx[i][j] += A[i][j] + B[i][j]

for g in res_mtx:
    print(*g)
```

**–£–º–Ω–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü üå∂Ô∏è**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—Ä–µ–º–Ω–æ–∂–∞–µ—Ç –¥–≤–µ –º–∞—Ç—Ä–∏—Ü—ã.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –ø–µ—Ä–≤–æ–π –º–∞—Ç—Ä–∏—Ü–µ, –∑–∞—Ç–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –ø–µ—Ä–≤–æ–π –º–∞—Ç—Ä–∏—Ü—ã, –∑–∞—Ç–µ–º –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞. –î–∞–ª–µ–µ —Å–ª–µ–¥—É—é—Ç —á–∏—Å–ª–∞ m –∏ k ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤—Ç–æ—Ä–æ–π –º–∞—Ç—Ä–∏—Ü—ã –∑–∞—Ç–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –≤—Ç–æ—Ä–æ–π –º–∞—Ç—Ä–∏—Ü—ã.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â—É—é –º–∞—Ç—Ä–∏—Ü—É, —Ä–∞–∑–¥–µ–ª—è—è —ç–ª–µ–º–µ–Ω—Ç—ã —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞.

```python
n, m = [int(i) for i in input().split()]  # —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–ª–±—Ü—ã –º–∞—Ç—Ä–∏—Ü—ã 1
A = [[int(i) for i in input().split()] for _ in range(n)]  # –º–∞—Ç—Ä–∏—Ü–∞ 1
null = input()  # –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ –º–µ–∂–¥—É –º–∞—Ç—Ä–∏—Ü–∞–º–∏
m, k = [int(i) for i in input().split()]  # —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–ª–±—Ü—ã –º–∞—Ç—Ä–∏—Ü—ã 2
B = [[int(i) for i in input().split()] for _ in range(m)]  # –º–∞—Ç—Ä–∏—Ü–∞ 2
C = [[0] * k for i in range(n)]  # —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –ø–æ—Å–ª–µ —É–º–Ω–æ–∂–µ–Ω–∏—è –ê*–í

for i in range(n):  # –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç—Ä–æ–∫ —Å–∞—Ç—Ä–∏—Ü—ã 1
    for j in range(k):  # –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç–æ–ª–±—Ü–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã 2
        sum_multipl = 0  # —Å—É–º–º–∞ –ø–æ–º–Ω–æ–∂–µ–Ω–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å—Ç—Ä–æ–∫–∏ –º–∞—Ç—Ä–∏—Ü—ã 1 —Å —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ —Å—Ç–æ–ª–±—Ü–∞ –º–∞—Ç—Ä–∏—Ü—ã 2 
        for g in range(m):  # –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç—Ä–æ–∫ = —Å—Ç–æ–ª–±—Ü–æ–≤ –æ–±–µ–∏—Ö –º–∞—Ç—Ä–∏—Ü
            sum_multipl += A[i][g] * B[g][j]  # —É–º–Ω–æ–∂–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –∏ —Å–∫–ª–∞–¥—ã–≤–∞–µ–º —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º 
        C[i][j] = sum_multipl  # –º–µ–Ω—è–µ–º –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏—Ä—É–±—â–µ–π –º–∞—Ç—Ä–∏—Ü–µ –Ω—É–ª–∏ –Ω–∞ —Å—É–º–º—É —É–º–Ω–æ–∂–µ–Ω–Ω—ã—Ö —ç–ª–µ–º–µ—Ç–æ–≤ —Å—Ç—Ä–æ–∫–∏ —Å —ç–ª–µ–º–µ–Ω—Ç–≤–º–∏ —Å—Ç–æ–±—Ü–æ–≤

for row in C:
    print(*row)
```

**–í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã –≤ —Å—Ç–µ–ø–µ–Ω—å üå∂Ô∏è**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤–æ–¥–∏—Ç –∫–≤–∞–¥—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –≤ m-—É—é —Å—Ç–µ–ø–µ–Ω—å.

```python
from copy import deepcopy  # –¥–µ–ª–∞–µ—Ç –∫–æ–ø–∏—é –º–∞—Ç—Ä–∏—Ü—ã —Å –Ω–æ–≤—ã–º –∏–Ω–¥–µ–∫—Å–æ–º –∏ –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –ø–µ—Ä–≤–æ–∏—Å—Ç–æ—á–Ω–∏–∫–∞
n = int(input())  # —Å—Ç–æ–ª–±—Ü—ã –∏ —Å—Ç—Ä–æ–∫–∏
a = [[int(i) for i in input().split()] for _ in range(n)]  # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Ç—Ä–∏—Ü—É
m = int(input())  # —Å—Ç–µ–ø–µ–Ω—å –¥–ª—è –≤–æ–∑–≤–µ–¥–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã
b = deepcopy(a)  # —Å–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –º–∞—Ç—Ä–∏—Ü—ã, —á—Ç–æ–± –Ω–∞ –Ω–µ–µ —É–º–Ω–æ–∂–∞—Ç—å m - –∫–æ–ª–∏—á—Å—Ç–≤–æ —Ä–∞–∑

for i in range(m - 1):  # —Å—Ç–µ–ø–µ–Ω—å —Ä–∞–∑ —É–º–Ω–æ–∂–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É –ë –Ω–∞ –Ω–∞—á–∞–ª—å–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –∞
    c = [[0] * n for i in range(n)]  # –∫–∞–∂–¥—ã–π —Ä–∞–∑ –ø–æ—Å–ª–µ –ø—Ä–æ—Ö–æ–¥–∞ –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ —Å –æ–¥–Ω–∏–º —Å—Ç–æ–±–æ–º
    for j in range(n):  # —Å—Ç—Ä–æ–∫–∏
        for k in range(n):  # —Å—Ç–æ–ª–±—Ü—ã
            for g in range(n):  # —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤
                c[j][k] += b[j][g] * a[g][k]  # —É–º–Ω–æ–∂–∞–µ–º —ç–ª–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫ –º–∞—Ç—Ä–∏—Ü—ã —å –Ω–∞ —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç–æ–ª–±—Ü–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã –∞
    b = deepcopy(c)  # –¥–µ–ª–∞–µ–º –∫–æ–ø–∏—é –º–∞—Ç—Ä–∏—Ü—ã –ø–æ—Å–ª–µ –ø—Ä–æ—Ö–æ–¥–∞ –ø–æ –≤—Å–µ–º —Å—Ç—Ä–æ–∫–∞–º –∏ —Å—Ç–æ–ª–±—Ü–∞
for row in c:
    print(*row)  # –≤—ã–≤–æ–¥–∏–º –∏–∑ —Å–ø–∏—Å–∫–∞ –ø–æ—Å—Ç—Ä–æ—á–Ω–æ
```

**–ö–∞–∂–¥—ã–π n-—ã–π —ç–ª–µ–º–µ–Ω—Ç**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è —Å–∏–º–≤–æ–ª—ã, –∏ —á–∏—Å–ª–æ n. –ò–∑ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞–∑–¥–µ–ª—è–µ—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ –ø–æ–¥—Å–ø–∏—Å–∫–∏ —Ç–∞–∫, —á—Ç–æ n –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç —Ä–∞–∑–Ω—ã–º –ø–æ–¥—Å–ø–∏—Å–∫–∞–º.

```python
s = input().split()  # a b c d e f g h i j k l m n
n = int(input())  # 3
res = []
for i in range(n):
    res.append(s[i::n])
print(res)
```

**–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –≤ –æ–±–ª–∞—Å—Ç–∏ 2**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ –∑–∞—à—Ç—Ä–∏—Ö–æ–≤–∞–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã (/_).

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)]
largest = mtx[0][n - 1]

for i in range(n):
    for j in range(n):
        if i + j + 1 >= n and mtx[i][j] > largest:
            largest = mtx[i][j]

print(largest)
```

**–¢—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É. –¢—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ ‚Äî –º–∞—Ç—Ä–∏—Ü–∞, –ø–æ–ª—É—á–µ–Ω–Ω–∞—è –∏–∑ –∏—Å—Ö–æ–¥–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –∑–∞–º–µ–Ω–æ–π —Å—Ç—Ä–æ–∫ –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã. –ó–∞–¥–∞—á—É –º–æ–∂–Ω–æ —Ä–µ—à–∏—Ç—å –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞. 

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)]
for i in range(n):
    for j in range(n):
        print(mtx[j][i], end=' ')
    print()
```

**–°–Ω–µ–∂–∏–Ω–∫–∞**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–µ—á–µ—Ç–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ón –∑–∞–ø–æ–ª–Ω–∏–≤ –µ—ë —Å–∏–º–≤–æ–ª–∞–º–∏ . . –ó–∞—Ç–µ–º –∑–∞–ø–æ–ª–Ω–∏—Ç–µ —Å–∏–º–≤–æ–ª–∞–º–∏ * —Å—Ä–µ–¥–Ω—é—é —Å—Ç—Ä–æ–∫—É –∏ —Å—Ç–æ–ª–±–µ—Ü –º–∞—Ç—Ä–∏—Ü—ã, –≥–ª–∞–≤–Ω—É—é –∏ –ø–æ–±–æ—á–Ω—É—é –¥–∏–∞–≥–æ–Ω–∞–ª—å –º–∞—Ç—Ä–∏—Ü—ã. –í—ã–≤–µ–¥–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –Ω–∞ —ç–∫—Ä–∞–Ω, —Ä–∞–∑–¥–µ–ª—è—è —ç–ª–µ–º–µ–Ω—Ç—ã –ø—Ä–æ–±–µ–ª–∞–º–∏.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–µ—á–µ—Ç–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n,(n‚â•3) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ.

```python
n = int(input())  # 5
s = [['.'] * n for i in range(n)]
for i in range(n):
    for j in range(n):
        if i == j or n - i - 1 == j or i == n // 2 or j == n // 2:
            s[i][j] = '*'
for g in s:
    print(*g)
```

**–°–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –ø—Ä–æ–≤–µ—Ä–∫–∏ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ YES, –µ—Å–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞, –∏ —Å–ª–æ–≤–æ NO –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

```python
n = int(input())
flag = 'YES'
mtx = [[int(i) for i in input().split()] for _ in range(n)]
for i in range(n):
    for j in range(n):
        if mtx[i][j] != mtx[n - j - 1][n - i - 1]:
            flag = 'NO'
print(flag)
```

**–õ–∞—Ç–∏–Ω—Å–∫–∏–π –∫–≤–∞–¥—Ä–∞—Ç üå∂Ô∏è**

–õ–∞—Ç–∏–Ω—Å–∫–∏–º –∫–≤–∞–¥—Ä–∞—Ç–æ–º –ø–æ—Ä—è–¥–∫–∞ n –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Ä–∞–∑–º–µ—Ä–æ–º n√ón, –∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∏ –∫–∞–∂–¥—ã–π —Å—Ç–æ–ª–±–µ—Ü –∫–æ—Ç–æ—Ä–æ–π —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—Å–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –ª–∞—Ç–∏–Ω—Å–∫–∏–º –∫–≤–∞–¥—Ä–∞—Ç–æ–º.

```python
n = int(input())
mtx = [[int(i) for i in input().split()] for _ in range(n)]
num = list(range(1, n + 1))
flag = 'YES'

for i in range(n):
    row_nums = sorted(mtx[i])
    col_nums = sorted([mtx[j][i] for j in range(n)])
    if row_nums != num or col_nums != num:
        flag = 'NO'
        break
        
print(flag)
```

**–•–æ–¥—ã —Ñ–µ—Ä–∑—è**

–ù–∞ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–µ 8√ó8 —Å—Ç–æ–∏—Ç —Ñ–µ—Ä–∑—å. –û—Ç–º–µ—Ç—å—Ç–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ñ–µ—Ä–∑—è –Ω–∞ –¥–æ—Å–∫–µ –∏ –≤—Å–µ –∫–ª–µ—Ç–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—å–µ—Ç —Ñ–µ—Ä–∑—å. –ö–ª–µ—Ç–∫—É, –≥–¥–µ —Å—Ç–æ–∏—Ç —Ñ–µ—Ä–∑—å, –æ—Ç–º–µ—Ç—å—Ç–µ –±—É–∫–≤–æ–π Q, –∫–ª–µ—Ç–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—å–µ—Ç —Ñ–µ—Ä–∑—å, –æ—Ç–º–µ—Ç—å—Ç–µ —Å–∏–º–≤–æ–ª–∞–º–∏ *, –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ –∑–∞–ø–æ–ª–Ω–∏—Ç–µ —Ç–æ—á–∫–∞–º–∏.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ñ–µ—Ä–∑—è –Ω–∞ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–µ –≤ —à–∞—Ö–º–∞—Ç–Ω–æ–π –Ω–æ—Ç–∞—Ü–∏–∏ (—Ç–æ –µ—Å—Ç—å –≤ –≤–∏–¥–µ e4, –≥–¥–µ —Å–Ω–∞—á–∞–ª–∞ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –Ω–æ–º–µ—Ä —Å—Ç–æ–ª–±—Ü–∞ (–±—É–∫–≤–∞ –æ—Ç a –¥–æ h, —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ), –∑–∞—Ç–µ–º –Ω–æ–º–µ—Ä —Å—Ç—Ä–æ–∫–∏ (—Ü–∏—Ñ—Ä–∞ –æ—Ç 1 –¥–æ 8, —Å–Ω–∏–∑—É –≤–≤–µ—Ä—Ö)).

```python
n = 8
chessboard = [['.'] * n for _ in range(n)]
location = input()  # c4
x = ord(location[0]) - ord('a')
y = abs(int(location[1]) - 8)
for i in range(n):
    for j in range(n):
        if (abs(x - i) == abs(y - j)) or (i == x) or j == y:
            chessboard[j][i] = '*'
            chessboard[y][x] = 'Q'
for row in chessboard:
    print(*row)
```

**–î–∏–∞–≥–æ–Ω–∞–ª–∏, –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–µ –≥–ª–∞–≤–Ω–æ–π**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ón –∏ –∑–∞–ø–æ–ª–Ω—è–µ—Ç –µ—ë –ø–æ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—Ä–∞–≤–∏–ª—É:
- –Ω–∞ –≥–ª–∞–≤–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –Ω–∞ –º–µ—Å—Ç–µ –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –¥–æ–ª–∂–Ω–æ —Å—Ç–æ—è—Ç—å —á–∏—Å–ª–æ 0;
- –Ω–∞ –¥–≤—É—Ö –¥–∏–∞–≥–æ–Ω–∞–ª—è—Ö, –ø—Ä–∏–ª–µ–≥–∞—é—â–∏—Ö –∫ –≥–ª–∞–≤–Ω–æ–π, ‚Äì —á–∏—Å–ª–æ 1;
- –Ω–∞ —Å–ª–µ–¥—É—é—â–∏—Ö –¥–≤—É—Ö –¥–∏–∞–≥–æ–Ω–∞–ª—è—Ö ‚Äì —á–∏—Å–ª–æ 2, –∏ —Ç.–¥.

```python
n = int(input())  # 5
mtx = [[0] * n for _ in range(n)]
for i in range(n):
    for j in range(n):
        for k in range(n):
            if i - j == k or j - i == k:
                mtx[i][j] = k
for row in mtx:
    print(*row)
```

**–í–µ—Ä—à–∏–Ω–∞ –ø–∞—Ä–∞–±–æ–ª—ã**

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–∞—Ä–∞–±–æ–ª—ã –∏–º–µ–µ—Ç –≤–∏–¥ y =ax^2+bx+c, –≥–¥–µ a‚â†0. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –≤–≤–µ–¥–µ–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º a,b,c –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –≤–µ—Ä—à–∏–Ω—É –ø–∞—Ä–∞–±–æ–ª—ã.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ—Ä—à–∏–Ω—ã –ø–∞—Ä–∞–±–æ–ª—ã y=ax^2+bx+c –∏–º–µ—é—Ç –≤–∏–¥ (-b/2a; (4ac-b^2)/4a)

```python
def vertex_par(a, b, c):  # –∏—Ç–æ–≥ —Ñ—É–Ω–∫—Ü–∏–∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è –≤ –≤–∏–¥–µ –∫–æ—Ä—Ç–µ–∂–∞
    x = -(b / (2 * a))
    y = (4 * a * c - b ** 2) / (4 * a)

    return x, y

vertex = vertex_par(int(input()), int(input()), int(input()))  # -2, 6, 1
print(vertex)
```

**–ö–æ–Ω–∫—É—Ä—Å–Ω—ã–π –æ—Ç–±–æ—Ä**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç —Å–ø–∏—Å–æ–∫ —Ö–æ—Ä–æ—à–∏—Å—Ç–æ–≤ –∏ –æ—Ç–ª–∏—á–Ω–∏–∫–æ–≤ –≤ –∫–ª–∞—Å—Å–µ.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n, –¥–∞–ª–µ–µ —Å–ª–µ–¥—É–µ—Ç n —Å—Ç—Ä–æ–∫ —Å —Ñ–∞–º–∏–ª–∏–µ–π —à–∫–æ–ª—å–Ω–∏–∫–∞ –∏ –µ–≥–æ –æ—Ü–µ–Ω–∫–æ–π –Ω–∞ –∫–∞–∂–¥–æ–π –∏–∑ –Ω–∏—Ö.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å–Ω–∞—á–∞–ª–∞ –≤—Å–µ –≤–≤–µ–¥—ë–Ω–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏ —Å —Ñ–∞–º–∏–ª–∏—è–º–∏ –∏ –æ—Ü–µ–Ω–∫–∞–º–∏ —É—á–µ–Ω–∏–∫–æ–≤ –≤ —Ç–æ–º –∂–µ –ø–æ—Ä—è–¥–∫–µ. –ó–∞—Ç–µ–º —Å–ª–µ–¥—É–µ—Ç –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞, –∞ –∑–∞—Ç–µ–º –≤—ã–≤–æ–¥—è—Ç—Å—è —Å—Ç—Ä–æ–∫–∏ —Å —Ñ–∞–º–∏–ª–∏—è–º–∏ –∏ –æ—Ü–µ–Ω–∫–∞–º–∏ —Ö–æ—Ä–æ—à–∏—Å—Ç–æ–≤ –∏ –æ—Ç–ª–∏—á–Ω–∏–∫–æ–≤ (–≤ —Ç–æ–º –∂–µ –ø–æ—Ä—è–¥–∫–µ).

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –û—Ü–µ–Ω–∫–∞ —É—á–µ–Ω–∏–∫–∞ ‚Äì —ç—Ç–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –æ—Ç 1 –¥–æ 5.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –≤ –∫–ª–∞—Å—Å–µ –µ—Å—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ö–æ—Ä–æ—à–∏—Å—Ç (–æ–±–ª–∞–¥–∞—Ç–µ–ª—å –æ—Ü–µ–Ω–∫–∏ 4) –∏–ª–∏ –æ—Ç–ª–∏—á–Ω–∏–∫ (–ø–æ–ª—É—á–∏–≤—à–∏–π 5).

```python
n = int(input())  # 3
lst = [input().split() for _ in range(n)]

for i in range(len(lst)):
    print(lst[i][0], lst[i][1])
print()  # –ø–µ—á–∞—Ç–∞–µ—Ç –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –∏ –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç —Ü–∏–∫–ª

for j in range(len(lst)):
    if lst[j][1] in '45':
        print(lst[j][0], lst[j][1])
```

**–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¢—Ä–∏–±–æ–Ω–∞—á—á–∏**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n –∏ –≤—ã–≤–æ–¥–∏—Ç –ø–µ—Ä–≤—ã–µ n —á–∏—Å–µ–ª –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –¢—Ä–∏–±–æ–Ω–∞—á—á–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n(n‚â§100) ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–ª–µ–Ω–æ–≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —á–ª–µ–Ω—ã –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –¢—Ä–∏–±–æ–Ω–∞—á—á–∏, –æ—Ç–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¢—Ä–∏–±–æ–Ω–∞—á—á–∏ ‚Äì –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –≥–¥–µ –∫–∞–∂–¥–æ–µ –ø–æ—Å–ª–µ–¥—É—é—â–µ–µ —á–∏—Å–ª–æ —è–≤–ª—è–µ—Ç—Å—è —Å—É–º–º–æ–π —Ç—Ä–µ—Ö –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö: 1,1,1,3,5,9,17,31,57,105, ‚Ä¶

```python
n = int(input())  # 10
f1, f2, f3 = 1, 1, 1
for i in range(n):
    print(f1, end=' ')
    f1, f2, f3 = f2, f3, f1 + f2 + f3
```

 **–¢–∏–º—É—Ä –∏ –µ–≥–æ –∫–æ–º–∞–Ω–¥–∞ üë®‚Äçüè´**

–ù–∞ –ª–µ—Ç–Ω–∏—Ö –∫–∞–Ω–∏–∫—É–ª–∞—Ö –¢–∏–º—É—Ä –∏ —É—á–µ–Ω–∏–∫–∏ –æ–Ω–ª–∞–π–Ω-—à–∫–æ–ª—ã BEEGEEK —Ä–µ—à–∏–ª–∏ –æ—Ç–¥–æ—Ö–Ω—É—Ç—å. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ n —É—á–µ–Ω–∏–∫–æ–≤ —à–∫–æ–ª—ã –ø–æ–µ—Ö–∞–ª–∏ –æ—Ç–¥—ã—Ö–∞—Ç—å –Ω–∞ –º–æ—Ä–µ, m —É—á–µ–Ω–∏–∫–æ–≤ —Å—ä–µ–∑–¥–∏–ª–∏ –≤ –¥–µ—Ä–µ–≤–Ω—é, –∞ k —É—á–µ–Ω–∏–∫–æ–≤ —Å—Ö–æ–¥–∏–ª–∏ –≤ –≥–æ—Ä—ã. –û–∫–∞–∑–∞–ª–æ—Å—å, —á—Ç–æ –∏ –≤ –¥–µ—Ä–µ–≤–Ω–µ, –∏ –Ω–∞ –º–æ—Ä–µ –±—ã–ª–∏ x —É—á–µ–Ω–∏–∫–æ–≤, –∞ –≤ –¥–µ—Ä–µ–≤–Ω–µ –∏ –≤ –≥–æ—Ä–∞—Ö ‚Äì y —É—á–µ–Ω–∏–∫–æ–≤. –ü–æ–±—ã–≤–∞—Ç—å –∏ –≤ –≥–æ—Ä–∞—Ö, –∏ –Ω–∞ –º–æ—Ä–µ –Ω–µ —É–¥–∞–ª–æ—Å—å –Ω–∏–∫–æ–º—É.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É—á–µ–Ω–∏–∫–æ–≤ –≤ —à–∫–æ–ª–µ, –µ—Å–ª–∏ –Ω–∏–∫—Ç–æ –Ω–µ —Å–º–æ–≥ –ø–æ—Å–µ—Ç–∏—Ç—å –≤—Å–µ —Ç—Ä–∏ –º–µ—Å—Ç–∞ —Å—Ä–∞–∑—É, –∞ z —É—á–µ–Ω–∏–∫–æ–≤ –ø–∏—Å–∞–ª–∏ –î–í–ò –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –¥–ª—è –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏—è –≤ –ú–ì–£ –∏ –Ω–∏–∫—É–¥–∞ –Ω–µ –µ–∑–¥–∏–ª–∏.
- n ‚Äì –≤—Å–µ —É—á–µ–Ω–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–µ—Ö–∞–ª–∏ –Ω–∞ –º–æ—Ä–µ, 
- m ‚Äì –≤—Å–µ —É—á–µ–Ω–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ —Å—ä–µ–∑–¥–∏–ª–∏ –≤ –¥–µ—Ä–µ–≤–Ω—é, 
- k ‚Äì –≤—Å–µ —É—á–µ–Ω–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ —Å—Ö–æ–¥–∏–ª–∏ –≤ –≥–æ—Ä—ã.

```python
n, m, k, x, y, z = (int(input()) for _ in range(6))  # –º–æ—Ä–µ, –¥–µ—Ä–µ–≤–Ω—è, –≥–æ—Ä—ã, –≤ –¥–µ—Ä–µ–≤–Ω–µ –∏ –Ω–∞ –º–æ—Ä–µ, –≤ –¥–µ—Ä–µ–≤–Ω–µ –∏ –≤ –≥–æ—Ä–∞—Ö, –Ω–∏–∫—É–¥–∞ –Ω–µ –µ–∑–¥–∏–ª–∏
print((n - x) + (m - x - y) + (k - y) + x + y + z)
```

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ.

```python
print(len(set(input())))  # 12345678910
```

**–ù–µ–ø–æ–≤—Ç–æ—Ä–∏–º—ã–µ —Ü–∏—Ñ—Ä—ã**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ —Ü–∏—Ñ—Ä. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –≤–µ—Ä–Ω–æ –ª–∏, —á—Ç–æ –≤ –µ–µ –∑–∞–ø–∏—Å–∏ –Ω–∏ –æ–¥–Ω–∞ –∏–∑ —Ü–∏—Ñ—Ä –Ω–µ –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è?

```python
s = input()  # 1093482
f = set(s)
if len(s) == len(f):
  print('YES')
else:
  print('NO')
```

**–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã 1**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –≤—ã–≤–æ–¥–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –∫–∞–∂–¥–æ–≥–æ —Å—á–∏—Ç–∞–Ω–Ω–æ–≥–æ —Å–ª–æ–≤–∞ –±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞.

```python
lst = []
for i in range(int(input())):
    lst.append(len(set(input().lower())))
print(*lst, sep='\n')
```

**–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã 2**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –≤—ã–≤–æ–¥–∞ –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –≤–æ –≤—Å–µ—Ö —Å—á–∏—Ç–∞–Ω–Ω—ã—Ö —Å–ª–æ–≤–∞—Ö –±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞.

```python
mn = set()
for _ in range(int(input())):
    st = input().lower()
    for i in st:
        mn.add(i)
print(len(mn))
```

**–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ –≤ —Ç–µ–∫—Å—Ç–µ**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Å–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ —Ç–µ–∫—Å—Ç–∞.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Å–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ –±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –°–ª–æ–≤–æ–º —Å—á–∏—Ç–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–µ–ø—Ä–æ–±–µ–ª—å–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤, –∏–¥—É—â–∏—Ö –ø–æ–¥—Ä—è–¥, —Å–ª–æ–≤–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –æ–¥–Ω–∏–º –∏–ª–∏ –±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –ø—Ä–æ–±–µ–ª–æ–≤.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ó–Ω–∞–∫–∞–º–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è .,;:-?! –ø—Ä–µ–Ω–µ–±—Ä–µ–≥–∞–µ–º.

```python
words = [word.lower().strip('.,;:-?!') for word in input().split()]
# Milk is white and so is glue, Ghosts are white and they say BOO!

print(len(set(words)))
```

**–í—Å—Ç—Ä–µ—á–∞–ª–æ—Å—å –ª–∏ —á–∏—Å–ª–æ —Ä–∞–Ω—å—à–µ?**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è —á–∏—Å–ª–∞. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ –≤—ã–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–æ YES (–Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ), –µ—Å–ª–∏ —ç—Ç–æ —á–∏—Å–ª–æ —Ä–∞–Ω–µ–µ –≤—Å—Ç—Ä–µ—á–∞–ª–æ—Å—å –≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –∏–ª–∏ NO –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í–µ–¥—É—â–∏–µ –Ω—É–ª–∏ –≤ —á–∏—Å–ª–∞—Ö –¥–æ–ª–∂–Ω—ã –∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞—Ç—å—Å—è.

```python
num = input().split()  # 1 1 2 2 5 5 5 5 6 7 8
plenty = set()
for c in num:
    if c.lstrip('0') in plenty:
        print('YES')
    else:
        plenty.add(c)
        print('NO')
```

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –Ω–∏–∫–Ω–µ–π–º—ã —Ä–∞–∑–Ω—ã—Ö —É—á–∞—â–∏—Ö—Å—è –Ω–µ –º–æ–≥—É—Ç —Å–æ–≤–ø–∞–¥–∞—Ç—å.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 3. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –Ω–∏–∫–Ω–µ–π–º —É—á–∞—â–µ–≥–æ—Å—è –Ω–µ –º–æ–∂–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–∏–º–≤–æ–ª –¥–≤–æ–µ—Ç–æ—á–∏—è (:).

```python
n = int(input())
m = set()  # –ú–Ω–æ–∂–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∫–æ—Ç–æ—Ä—ã–µ –≤–µ—Ä–Ω–æ —Ä–µ—à–∏–ª–∏
base_cnt = 0  # –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ—à–µ–Ω–∏–π = 0
cnt = 0  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä–Ω—ã—Ö —Ä–µ—à–µ–Ω–∏–π = 0
for i in range(n):
    s = input().split(': ')  # —Å–ø–ª–∏—Ç –ø–æ –¥–≤–æ–µ—Ç–æ—á–∏—é —Å –ø—Ä–æ–±–µ–ª–æ–º
    base_cnt += 1  # –ö –æ–±—â–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ä–µ—à–µ–Ω–∏–π –ø—Ä–∏–±–∞–≤–ª—è–µ–º 1
    if 'Correct' in s:
        m.add(str(s))  # —Ç–æ –≤ –º–Ω–æ–∂–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–º—ë–Ω –¥–æ–±–∞–≤–ª—è–µ–º –∏–º—è 
        cnt += 1  # –∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≤–µ—Ä–Ω—ã—Ö —Ä–µ—à–µ–Ω–∏–π –ø—Ä–∏–±–∞–≤–ª—è–µ–º 1. –ù–µ–≤–µ—Ä–Ω—ã–µ —Ä–µ—à–µ–Ω–∏—è –±—É–¥–µ–º –ø—Ä–æ—Å—Ç–æ –∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞—Ç—å
if cnt == 0:
    print('–í—ã –º–æ–∂–µ—Ç–µ —Å—Ç–∞—Ç—å –ø–µ—Ä–≤—ã–º, –∫—Ç–æ —Ä–µ—à–∏—Ç —ç—Ç—É –∑–∞–¥–∞—á—É')
else:
    print('–í–µ—Ä–Ω–æ —Ä–µ—à–∏–ª–∏', len(m), '—É—á–∞—â–∏—Ö—Å—è')
    print('–ò–∑ –≤—Å–µ—Ö –ø–æ–ø—ã—Ç–æ–∫ ', int((cnt / base_cnt) * 100 + 0.5), '%', ' –≤–µ—Ä–Ω—ã—Ö', sep='')
```

**–û–±—â–∏–µ —á–∏—Å–ª–∞**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Ç–µ–∫—Å—Ç–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ —á–∏—Å–ª–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –≤—Å–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –µ—Å—Ç—å –∫–∞–∫ –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ, —Ç–∞–∫ –∏ –≤–æ –≤—Ç–æ—Ä–æ–π.

```python
a = set(int(i) for i in input().split())  # 1 2 3
b = set(int(i) for i in input().split())  # 1 2 4 5

print(*sorted(a & b))
```

**–û–±—â–∏–µ —Ü–∏—Ñ—Ä—ã**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n, –∞ –∑–∞—Ç–µ–º n —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –≤—Å–µ –æ–±—â–∏–µ —Ü–∏—Ñ—Ä—ã –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è —É –≤—Å–µ—Ö –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª.

```python
n = int(input())  # 3
myset = set(input())  # 4545436
for _ in range(n - 1):
    myset.intersection_update(set(input()))  # 345345, 76857
print(*(sorted(myset)))
```

**–û–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ü–∏—Ñ—Ä—ã ‚öñÔ∏è**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é, –µ—Å—Ç—å –ª–∏ –≤ –¥–∞–Ω–Ω—ã—Ö —á–∏—Å–ª–∞—Ö –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ü–∏—Ñ—Ä—ã.

```python
print(("YES", "NO")[set(input()).isdisjoint(input())])  # 114, 223
```

**–£—Ä–æ–∫ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∏ üë®‚Äçüíª**

–î–∞–Ω—ã –ø–æ 10-–±–∞–ª–ª—å–Ω–æ–π —à–∫–∞–ª–µ –æ—Ü–µ–Ω–∫–∏ –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–µ —Ç—Ä–µ—Ö —É—á–µ–Ω–∏–∫–æ–≤. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã–≤–æ–¥—è—â—É—é –º–Ω–æ–∂–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–æ–∫, –∫–æ—Ç–æ—Ä—ã–µ –µ—Å—Ç—å –∏ —É –ø–µ—Ä–≤–æ–≥–æ, –∏ —É –≤—Ç–æ—Ä–æ–≥–æ —É—á–µ–Ω–∏–∫–æ–≤, –Ω–æ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç —É —Ç—Ä–µ—Ç—å–µ–≥–æ —É—á–µ–Ω–∏–∫–∞. –û—Ü–µ–Ω–∫–∞ —É—á–µ–Ω–∏–∫–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 0 –¥–æ 10 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

```python
a, b, c = (set(int(i) for i in input().split()) for i in range(3))
# 1 5 4 2 5 6 6 2 3 3 5 2
# 2 3 5 1 2 1 2 6 7 1 1 6
# 1 4 6 9 8 7 0 9 0 9 8 10
print(*sorted(set(a.intersection(b).difference(c)))[::-1])
```

**–£—Ä–æ–∫ —Ñ–∏–∑–∏–∫–∏ üß≤**

–î–∞–Ω—ã –ø–æ 10-–±–∞–ª–ª—å–Ω–æ–π —à–∫–∞–ª–µ –æ—Ü–µ–Ω–∫–∏ –ø–æ —Ñ–∏–∑–∏–∫–µ —Ç—Ä–µ—Ö —É—á–µ–Ω–∏–∫–æ–≤. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–æ–∫ —Ç—Ä–µ—Ç—å–µ–≥–æ —É—á–µ–Ω–∏–∫–∞, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –Ω–∏ —É –ø–µ—Ä–≤–æ–≥–æ, –Ω–∏ —É –≤—Ç–æ—Ä–æ–≥–æ —É—á–µ–Ω–∏–∫–∞.

```python
set_1, set_2, set_3 = (set(input().split()) for _ in range(3))
# 1 5 4 2 5 6 6 2 3 3 5 2
# 2 3 5 1 2 1 2 6 7 1 1 6
# 1 4 6 9 8 7 0 9 0 9 8 10
print(*(sorted(set_3 - (set_1 | set_2), key=int, reverse=True)))
```

**–£—Ä–æ–∫ –±–∏–æ–ª–æ–≥–∏–∏ üå±**

–î–∞–Ω—ã –ø–æ 10-–±–∞–ª–ª—å–Ω–æ–π —à–∫–∞–ª–µ –æ—Ü–µ–Ω–∫–∏ –ø–æ –±–∏–æ–ª–æ–≥–∏–∏ —Ç—Ä–µ—Ö —É—á–µ–Ω–∏–∫–æ–≤. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–æ–∫, –Ω–µ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏—Ö—Å—è –Ω–∏ —É –æ–¥–Ω–æ–≥–æ –∏–∑ —Ç—Ä–µ—Ö —É—á–µ–Ω–∏–∫–æ–≤.

```python
set1 = set(int(i) for i in input().split())  # 1 5 4 2 5 6 6 2 3 3 5 2
set2 = set(int(i) for i in input().split())  # 2 3 5 1 2 1 2 6 7 1 1 6
set3 = set(int(i) for i in input().split())  # 1 4 6 8 8 7 0 6 0 3 8 1

print(*sorted(set(range(11)) - set1 - set2 - set3))
```

**–î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ üìù**

–£—á–∏—Ç–µ–ª—å –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ –≤ –∫–ª–∞—Å—Å–µ –∏ –ø–æ–ª—É—á–∏–ª —Å–ª–µ–¥—É—é—â–∏–µ –æ—Ç–≤–µ—Ç—ã: –∏–∑ n —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ —É m –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ —Å—ä–µ–ª–∞ —Å–æ–±–∞–∫–∞, —É k –æ—Ç–∫–ª—é—á–∏–ª–∏ —Å–≤–µ—Ç, –∞ p —É—á–µ–Ω–∏–∫–æ–≤ –ø–æ—Å—Ç–∏–≥–ª–∏ –æ–±–∞ –Ω–µ—Å—á–∞—Å—Ç—å—è. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —Å–∫–æ–ª—å–∫–æ —á–µ–ª–æ–≤–µ–∫ –≤—ã–ø–æ–ª–Ω–∏–ª–æ –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ.

```python
n,m,k,p = [int(input()) for i in range(4)]
res = n - m - k + p
print(res)
```

**–í–æ—Å—Ö–æ–¥ üõ∞Ô∏è**

–ù–∞ —Å–ø—É—Ç–Ω–∏–∫–µ ¬´–í–æ—Å—Ö–æ–¥¬ª —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –ø—Ä–∏–±–æ—Ä –¥–ª—è –∏–∑–º–µ—Ä–µ–Ω–∏—è —Å–æ–ª–Ω–µ—á–Ω–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏. –ö–∞–∂–¥—É—é –º–∏–Ω—É—Ç—É –æ–Ω –ø–µ—Ä–µ–¥–∞—ë—Ç –≤ –æ–±—Å–µ—Ä–≤–∞—Ç–æ—Ä–∏—é –ø–æ –∫–∞–Ω–∞–ª—É —Å–≤—è–∑–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–Ω–µ—Ä–≥–∏–∏ —Å–æ–ª–Ω–µ—á–Ω–æ–≥–æ –∏–∑–ª—É—á–µ–Ω–∏—è. –î–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –Ω–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –¥–µ—Ä–∂–∞—Ç—å –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –¥–∞–Ω–Ω—ã–µ, –ø–æ—ç—Ç–æ–º—É –∏—Ö –Ω—É–∂–Ω–æ —É–¥–∞–ª–∏—Ç—å. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫–∞–∑–∞–Ω–∏–π —Å–ø—É—Ç–Ω–∏–∫–∞, –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º 1 —Ç–µ—Å—Ç: —É –Ω–∞—Å –ø–æ–¥–∞—é—Ç—Å—è –¥–∞–Ω–Ω—ã–µ 10 20 30 10 40 10. –ú—ã –≤–∏–¥–∏–º, —á—Ç–æ —á–∏—Å–ª–æ 10 —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è —Ç—É—Ç 3 —Ä–∞–∑–∞ ‚Äì –∑–Ω–∞—á–∏—Ç, –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è —á–∏—Å–ª–∞ 10 –Ω–∞–¥–æ —É–¥–∞–ª–∏—Ç—å. –¢–∞–∫–∏—Ö —É –Ω–∞—Å 2 (–ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ 10 –º—ã –Ω–µ —É—á–∏—Ç—ã–≤–∞–µ–º). –î—Ä—É–≥–∏–µ —á–∏—Å–ª–∞ –Ω–µ –ø–æ–≤—Ç–æ—Ä—è—é—Ç—Å—è, –ø–æ—ç—Ç–æ–º—É –æ—Ç–≤–µ—Ç –±—É–¥–µ—Ç 2.

```python
s = [int(i) for i in input().split()]  # 10 20 30 10 40 10
m = set(s)
res = len(s) - len(m)
print(res)
```

**–ì–æ—Ä–æ–¥–∞ üåÉ**

–¢–∏–º—É—Ä –∏ –†—É—Å–ª–∞–Ω –∏–≥—Ä–∞—é—Ç –≤ –∏–≥—Ä—É –≥–æ—Ä–æ–¥–∞. –û–Ω–∏ –æ—á–µ–Ω—å –ª—é–±—è—Ç —ç—Ç—É –∏–≥—Ä—É –∏ –∑–Ω–∞—é—Ç –º–Ω–æ–≥–æ –≥–æ—Ä–æ–¥–æ–≤, –æ—Å–æ–±–µ–Ω–Ω–æ –¢–∏–º—É—Ä, –æ–¥–Ω–∞–∫–æ –∫ –∫–æ–Ω—Ü—É –∏–≥—Ä—ã –≤–≤–∏–¥—É —Å–≤–æ–µ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞ –∑–∞–±—ã–≤–∞—é—Ç, –∫–∞–∫–∏–µ –≥–æ—Ä–æ–¥–∞ —É–∂–µ –Ω–∞–∑—ã–≤–∞–ª–∏.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å—á–∏—Ç—ã–≤–∞—é—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –∏–≥—Ä–µ –∏ —Å–æ–æ–±—â–∞—é—â—É—é —Ä–µ–±—è—Ç–∞–º, —á—Ç–æ –æ—á–µ—Ä–µ–¥–Ω–æ–π –≥–æ—Ä–æ–¥ –Ω–∞–∑–≤–∞–Ω –ø–æ–≤—Ç–æ—Ä–Ω–æ.

```python
n = int(input())
s = {input() for _ in range(n)}
if input() in s:
    print('REPEAT')
else:
    print('OK')
```

**–ö–Ω–∏–≥–∏ –Ω–∞ –ø—Ä–æ—á—Ç–µ–Ω–∏–µ üìö**

–†—É—Å–ª–∞–Ω –ø–æ–ª—É—á–∏–ª –≤ –∫–æ–Ω—Ü–µ —É—á–µ–±–Ω–æ–≥–æ –≥–æ–¥–∞ —Å–ø–∏—Å–æ–∫ –ª–∏—Ç–µ—Ä–∞—Ç—É—Ä—ã –Ω–∞ –ª–µ—Ç–æ. –¢–µ–ø–µ—Ä—å –µ–º—É –Ω–∞–¥–æ –≤—ã—è—Å–Ω–∏—Ç—å, –∫–∞–∫–∏–µ –∫–Ω–∏–≥–∏ –∏–∑ —ç—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞ —É –Ω–µ–≥–æ –µ—Å—Ç—å. –£ –†—É—Å–ª–∞–Ω–∞ –Ω–∞ –∫–æ–º–ø—å—é—Ç–µ—Ä–µ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ñ–∞–π–ª–µ –∑–∞–ø–∏—Å–∞–Ω—ã –≤—Å–µ –∫–Ω–∏–≥–∏ –∏–∑ –µ–≥–æ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –¥–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏ –∏–∑ —Å–ø–∏—Å–∫–∞ –Ω–∞ –ø—Ä–æ—á—Ç–µ–Ω–∏–µ, –µ—Å—Ç—å –æ–Ω–∞ —É –†—É—Å–ª–∞–Ω–∞ –∏–ª–∏ –Ω–µ—Ç.

```python
m, n = int(input()), int(input())
lib = [input() for _ in range(m)]
hw = [input() for _ in range(n)]
for i in range(n):
    if hw[i] in lib:
        print('YES')
    else:
        print('NO')
```

**–°—Ç—Ä–∞–Ω–Ω–æ–µ —É–≤–ª–µ—á–µ–Ω–∏–µ üìÑ**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –æ–±—â–∏–µ —á–∏—Å–ª–∞ –¥–≤—É—Ö –ª–∏—Å—Ç–æ—á–∫–æ–≤ –∏–ª–∏ —Å–æ–æ–±—â–∞–µ—Ç, —á—Ç–æ –¥–µ–Ω—å –Ω–µ —É–¥–∞–ª—Å—è. üòè

```python
set1 = {int(i) for i in input().split()}  # 6 56 7 34 14
set2 = {int(i) for i in input().split()}  # 675 45 246 2 1

if set1.isdisjoint(set2):
    print('BAD DAY')
else:
    print(*sorted(set1 & set2, reverse=True))
```

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –ø–æ–∑–≤–æ–ª—è—é—â—É—é —Ä—É–∫–æ–≤–æ–¥–∏—Ç–µ–ª—é –≤—ã—è—Å–Ω–∏—Ç—å, —Å–∫–æ–ª—å–∫–æ —É—á–µ–Ω–∏–∫–æ–≤ –∏–∑—É—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç.

**–û–Ω–ª–∞–π–Ω-—à–∫–æ–ª–∞ BEEGEEK 5 üêùüå∂Ô∏è**

–ö–∞–∂–¥—ã–π —É—á–µ–Ω–∏–∫, –æ–±—É—á–∞—é—â–∏–π—Å—è –≤ –æ–Ω–ª–∞–π–Ω-—à–∫–æ–ª–µ BEEGEEK –∏–∑—É—á–∞–µ—Ç –ª–∏–±–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫—É, –ª–∏–±–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É, –ª–∏–±–æ –æ–±–∞ —ç—Ç–∏—Ö –ø—Ä–µ–¥–º–µ—Ç–∞. –£ —Ä—É–∫–æ–≤–æ–¥–∏—Ç–µ–ª—è —à–∫–æ–ª—ã –µ—Å—Ç—å —Å–ø–∏—Å–∫–∏ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –∫–∞–∂–¥—ã–π –ø—Ä–µ–¥–º–µ—Ç. –°–ª—É—á–∞–π–Ω–æ —Å–ø–∏—Å–∫–∏ –≤—Å–µ—Ö —É—á–µ–Ω–∏–∫–æ–≤ –ø–µ—Ä–µ–º–µ—à–∞–ª–∏—Å—å.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª–∏—Ç —Ä—É–∫–æ–≤–æ–¥–∏—Ç–µ–ª—é –≤—ã—è—Å–Ω–∏—Ç—å, —Å–∫–æ–ª—å–∫–æ —É—á–µ–Ω–∏–∫–æ–≤ –∏–∑—É—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç.

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—Ä–µ–¥–∏ —É—á–µ–Ω–∏–∫–æ–≤ —à–∫–æ–ª—ã BEEGEEK –Ω–µ—Ç –æ–¥–Ω–æ—Ñ–∞–º–∏–ª—å—Ü–µ–≤.

```python
m, n = [int(input()) for _ in range(2)]

# —Å–ø–∏—Å–æ–∫ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ (–≤ —Ç–æ–º —á–∏—Å–ª–µ —Å –ø–æ–≤—Ç–æ—Ä–∞–º–∏)
students_list = [input() for _ in range(m + n)]
# –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ (—É–∂–µ –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–æ–≤)
students_set = set(students_list)

# –∏–∑—É—á–∞—é—â–∏–µ —Å—Ä–∞–∑—É –¥–≤–∞ –ø—Ä–µ–¥–º–µ—Ç–∞
both_subjects = len(students_list) - len(students_set)
# –∏–∑—É—á–∞—é—â–∏–µ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç
only_one_subject = len(students_set) - both_subjects

print(only_one_subject or "NO")
```

**–û–Ω–ª–∞–π–Ω-—à–∫–æ–ª–∞ BEEGEEK 6 üêùüå∂Ô∏è**

–†—É–∫–æ–≤–æ–¥–∏—Ç–µ–ª—é –æ–Ω–ª–∞–π–Ω-—à–∫–æ–ª—ã BEEGEEK –∑–∞—Ö–æ—Ç–µ–ª–æ—Å—å —É–∑–Ω–∞—Ç—å, –∫—Ç–æ –∏–∑ –µ–≥–æ —É—á–µ–Ω–∏–∫–æ–≤ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞–ª –Ω–∞ –≤—Å–µ—Ö —É—Ä–æ–∫–∞—Ö —Å –Ω–∞—á–∞–ª–∞ —É—á–µ–±–Ω–æ–≥–æ –≥–æ–¥–∞. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —É—Ä–æ–∫–∞ –µ—Å—Ç—å –ª–∏—Å—Ç–æ–∫ —Å–æ —Å–ø–∏—Å–∫–æ–º –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞–≤—à–∏—Ö —É—á–µ–Ω–∏–∫–æ–≤.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é —Ñ–∞–º–∏–ª–∏–∏ —É—á–µ–Ω–∏–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞–ª–∏ –Ω–∞ –≤—Å–µ—Ö —É—Ä–æ–∫–∞—Ö.
